## Map xCO2 data location first
Create Chloropleth map with Folium

```python
m = folium.Map(location=[27.6648, -81.5158],
               tiles='cartodbpositron',
               zoom_start=6)
```

```python
xco2.apply(lambda row: folium.Circle(location=[row['Lat'], row['Long']], radius=2, color='crimson').add_to(m), axis=1)
```

```python
m
```

```python
# Narrow the long range of dataset; to just around Orlando
# Choosing LatLong was done by trial and error

xco2_red = xco2[(xco2['Long'] < -80.5) & (xco2['Long'] > -82) & (xco2['Lat'] < 30)]
xco2_red.shape
```

```python
from branca.element import Figure
fig1=Figure(width=1000, height=700)

m1 = folium.Map(location=[27.6648, -81.5158],
               zoom_start=6)

xco2_red.apply(lambda row: folium.Circle(location=[row['Lat'], row['Long']], radius=2, color='crimson').add_to(m1), axis=1)

fig1.add_child(m1)
folium.TileLayer('OpenStreetMap').add_to(m1)
folium.TileLayer('cartodbdark_matter').add_to(m1)
folium.LayerControl().add_to(m1)

```

```python
from folium import plugins
from folium.plugins import HeatMap

m2 = folium.Map(location=[27.6648, -81.5158],
               zoom_start=6)

HeatMap(data=xco2_red[['Lat', 'Long', 'xCO2']].groupby(['Lat','Long']).mean().reset_index().values.tolist(), 
        radius=8, max_zoom=10).add_to(m2)

m2
```
